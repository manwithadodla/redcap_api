```python
import pandas as pd
import numpy as np
import redcap

API_Token = ''
project = redcap.Project('https://redcap.prc.utexas.edu/redcap/api/', API_Token)

#set these variables
data_csv_path = '/Users/md35727/Downloads/data (5)/data.csv'
data_folder_path = '/Users/md35727/Downloads/data (5)/'
participant_identifier_var = "participant_info:1"

participant_info = pd.read_csv(data_csv_path, index_col= False)
```

```python
for i in range(len(participant_info.index)):
    duplicate_participant = False
    participant_id = participant_info.loc[i, participant_identifier_var]
    textfile = participant_info.loc[i, "little_man_experiment:1"]
    subset = project.export_records(fields=['participant_id', 'little_man_lmt_complete'])
    for s in subset:
        if (str(s['participant_id']) == str(participant_id) and s['little_man_lmt_complete'] == '1'):
            duplicate_participant = True
            print('Participant ' + str(participant_id) + ' is already in REDCap.')
    if not duplicate_participant:
        with open(data_folder_path + textfile) as myfile:
            lines = list(myfile)
            lines_list = lines[-6:]

            line1 = lines_list[0].split(" ")
            num_correct = (line1[0])
            num_wrong = (line1[1])
            num_timedout = (line1[2])
            avg_rt = (line1[3])
            correct_rt = (line1[4])  
            wrong_rt = (line1[5].split('\n')[0])
            percent_correct = int(num_correct)/32
            percent_wrong = int(num_wrong)/32
            lmt_efficiency = percent_correct/int(correct_rt)

            line2 = lines_list[1].split(" ")
            lat_correct_BDL = (line2[0])
            lat_correct_BDR = (line2[1])
            lat_correct_BUL = (line2[2])
            lat_correct_BUR = (line2[3])
            lat_correct_FDL = (line2[4])  
            lat_correct_FDR = (line2[5])  
            lat_correct_FUL = (line2[6])  
            lat_correct_FUR = (line2[7].split('\n')[0])

            line3 = lines_list[2].split(" ")
            num_correct_BDL = (line3[0])
            num_correct_BDR = (line3[1])
            num_correct_BUL = (line3[2])
            num_correct_BUR = (line3[3])
            num_correct_FDL = (line3[4])  
            num_correct_FDR = (line3[5])  
            num_correct_FUL = (line3[6])  
            num_correct_FUR = (line3[7].split('\n')[0])

            line4 = lines_list[3].split(" ")
            lat_wrong_BDL = (line4[0])
            lat_wrong_BDR = (line4[1])
            lat_wrong_BUL = (line4[2])
            lat_wrong_BUR = (line4[3])
            lat_wrong_FDL = (line4[4])  
            lat_wrong_FDR = (line4[5])  
            lat_wrong_FUL = (line4[6])  
            lat_wrong_FUR = (line4[7].split('\n')[0])

            line5 = lines_list[4].split(" ")
            num_wrong_BDL = (line5[0])
            num_wrong_BDR = (line5[1])
            num_wrong_BUL = (line5[2])
            num_wrong_BUR = (line5[3])
            num_wrong_FDL = (line5[4])  
            num_wrong_FDR = (line5[5])  
            num_wrong_FUL = (line5[6])  
            num_wrong_FUR = (line5[7].split('\n')[0])

            line6 = lines_list[5].split(" ")
            num_timedout_BDL = (line6[0])
            num_timedout_BDR = (line6[1])
            num_timedout_BUL = (line6[2])
            num_timedout_BUR = (line6[3])
            num_timedout_FDL = (line6[4])  
            num_timedout_FDR = (line6[5])  
            num_timedout_FUL = (line6[6])  
            num_timedout_FUR = (line6[7].split('\n')[0])
            data = [{'participant_id': int(participant_id), 'lmt_01': percent_correct, 'lmt_02': percent_wrong, 'lmt_03': num_correct, 'lmt_04': num_wrong,
                     'lmt_05': num_timedout, 'lmt_06': avg_rt, 'lmt_07': correct_rt, 'lmt_08': wrong_rt, 'lmt_09': lmt_efficiency, 
                     'lmt_10': lat_correct_BDL, 'lmt_11': lat_correct_BDR, 'lmt_12': lat_correct_BUL, 'lmt_13': lat_correct_BUR, 
                     'lmt_14': lat_correct_FDL, 'lmt_15': lat_correct_FDR, 'lmt_16': lat_correct_FUL, 'lmt_17': lat_correct_FUR, 
                     'lmt_18': num_correct_BDL, 'lmt_19': num_correct_BDR, 'lmt_20': num_correct_BUL, 'lmt_21': num_correct_BUR, 
                     'lmt_22': num_correct_FDL, 'lmt_23': num_correct_FDR, 'lmt_24': num_correct_FUL, 'lmt_25': num_correct_FUR,
                     'lmt_26': lat_wrong_BDL, 'lmt_27': lat_wrong_BDR, 'lmt_28': lat_wrong_BUL, 'lmt_29': lat_wrong_BUR, 
                     'lmt_30': lat_wrong_FDL, 'lmt_31': lat_wrong_FDR, 'lmt_32': lat_wrong_FUL, 'lmt_33': lat_wrong_FUR, 
                     'lmt_34': num_wrong_BDL, 'lmt_35': num_wrong_BDR, 'lmt_36': num_wrong_BUL, 'lmt_37': num_wrong_BUR, 
                     'lmt_38': num_wrong_FDL, 'lmt_39': num_wrong_FDR, 'lmt_40': num_wrong_FUL, 'lmt_41': num_wrong_FUR,
                     'lmt_42': num_timedout_BDL, 'lmt_43': num_timedout_BDR, 'lmt_44': num_timedout_BUL, 'lmt_45': num_timedout_BUR, 
                     'lmt_46': num_timedout_FDL, 'lmt_47': num_timedout_FDR, 'lmt_48': num_timedout_FUL, 'lmt_49': num_timedout_FUR,
                     'little_man_lmt_complete': 1}]
            import_data = project.import_records(data)
            print('Participant ' + str(participant_id) + ' data has been added to REDCap.')

```
