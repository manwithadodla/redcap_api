```python
import pandas as pd
import numpy as np
import redcap

API_Token = ''
project = redcap.Project('https://redcap.prc.utexas.edu/redcap/api/', API_Token)

data_csv_path = '/Users/md35727/Downloads/data (15)/data.csv'
data_folder_path = '/Users/md35727/Downloads/data (15)/'
participant_identifier_var = "participant_info:1"

participant_info = pd.read_csv(data_csv_path, index_col= False)
```

```python
#looping through indicies in data excel
for i in range(len(participant_info.index)):
    duplicate_participant = False
    participant_id = participant_info.loc[i, participant_identifier_var]
    textfile = participant_info.loc[i, "nback_task:1"]
    subset = project.export_records(fields=['participant_id', 'nback_nbt_complete'])
    for s in subset:
        if (str(s['participant_id']) == str(participant_id) and s['nback_nbt_complete'] == '1'):
            duplicate_participant = True
            print('Participant ' + str(participant_id) + ' is already in REDCap.')
    if not duplicate_participant:
        if (type(textfile) == str):
            print(textfile)
            with open(data_folder_path + textfile) as myfile:
                lines = list(myfile)
                lines_list = []
                for line in lines:
                    if len(line.split(" ")) == 5:
                        lines_list.append(line)

                b1_1n = lines_list[0].split(" ")
                b1_hits_1n = (b1_1n[0])
                b1_false_alarms_1n = (b1_1n[1])
                b1_misses_1n = (b1_1n[2])
                b1_correct_reject_1n = (b1_1n[3])
                b1_hits_rt_1n = (b1_1n[4].split('\n')[0])

                b2_1n = lines_list[1].split(" ")
                b2_hits_1n = (b2_1n[0])
                b2_false_alarms_1n = (b2_1n[1])
                b2_misses_1n = (b2_1n[2])
                b2_correct_reject_1n = (b2_1n[3])
                b2_hits_rt_1n = (b2_1n[4].split('\n')[0])

                b3_1n = lines_list[2].split(" ")
                b3_hits_1n = (b3_1n[0])
                b3_false_alarms_1n = (b3_1n[1])
                b3_misses_1n = (b3_1n[2])
                b3_correct_reject_1n = (b3_1n[3])
                b3_hits_rt_1n = (b3_1n[4].split('\n')[0])

                b1_2n = lines_list[3].split(" ")
                b1_hits_2n = (b1_2n[0])
                b1_false_alarms_2n = (b1_2n[1])
                b1_misses_2n = (b1_2n[2])
                b1_correct_reject_2n = (b1_2n[3])
                b1_hits_rt_2n = (b1_2n[4].split('\n')[0])

                b2_2n = lines_list[4].split(" ")
                b2_hits_2n = (b2_2n[0])
                b2_false_alarms_2n = (b2_2n[1])
                b2_misses_2n = (b2_2n[2])
                b2_correct_reject_2n = (b2_2n[3])
                b2_hits_rt_2n = (b2_2n[4].split('\n')[0])

                b3_2n = lines_list[5].split(" ")
                b3_hits_2n = (b3_2n[0])
                b3_false_alarms_2n = (b3_2n[1])
                b3_misses_2n = (b3_2n[2])
                b3_correct_reject_2n = (b3_2n[3])
                b3_hits_rt_2n = (b3_2n[4].split('\n')[0])

                b1_3n = lines_list[6].split(" ")
                b1_hits_3n = (b1_3n[0])
                b1_false_alarms_3n = (b1_3n[1])
                b1_misses_3n = (b1_3n[2])
                b1_correct_reject_3n = (b1_3n[3])
                b1_hits_rt_3n = (b1_3n[4].split('\n')[0])

                b2_3n = lines_list[7].split(" ")
                b2_hits_3n = (b2_3n[0])
                b2_false_alarms_3n = (b2_3n[1])
                b2_misses_3n = (b2_3n[2])
                b2_correct_reject_3n = (b2_3n[3])
                b2_hits_rt_3n = (b2_3n[4].split('\n')[0])

                b3_3n = lines_list[8].split(" ")
                b3_hits_3n = (b3_3n[0])
                b3_false_alarms_3n = (b3_3n[1])
                b3_misses_3n = (b3_3n[2])
                b3_correct_reject_3n = (b3_3n[3])
                b3_hits_rt_3n = (b3_3n[4].split('\n')[0])


            data = [{'participant_id': int(participant_id),

                     'nbt_b1_1back_hits': b1_hits_1n, 'nbt_b2_1back_hits': b2_hits_1n, 
                     'nbt_b3_1back_hits': b3_hits_1n,

                     'nbt_b1_1back_false_alarms': b1_false_alarms_1n, 'nbt_b2_1back_false_alarms': b2_false_alarms_1n,
                     'nbt_b3_1back_false_alarms': b3_false_alarms_1n, 

                     'nbt_b1_1back_misses': b1_misses_1n, 'nbt_b2_1back_misses': b2_misses_1n, 
                     'nbt_b3_1back_misses': b3_misses_1n,  

                     'nbt_b1_1back_correct_rejec': b1_correct_reject_1n, 'nbt_b2_1back_correct_rejec': b2_correct_reject_1n,
                     'nbt_b3_1back_correct_rejec': b3_correct_reject_1n,

                     'nbt_b1_1back_hits_rt': b1_hits_rt_1n, 'nbt_b2_1back_hits_rt': b2_hits_rt_1n,
                     'nbt_b3_1back_hits_rt': b3_hits_rt_1n,

                     'nbt_b1_2back_hits': b1_hits_2n, 'nbt_b2_2back_hits': b2_hits_2n, 
                     'nbt_b3_2back_hits': b3_hits_2n,  

                     'nbt_b1_2back_false_alarms': b1_false_alarms_2n, 'nbt_b2_2back_false_alarms': b2_false_alarms_2n,
                     'nbt_b3_2back_false_alarms': b3_false_alarms_2n, 

                     'nbt_b1_2back_misses': b1_misses_2n, 'nbt_b2_2back_misses': b2_misses_2n, 
                     'nbt_b3_2back_misses': b3_misses_2n,

                     'nbt_b1_2back_correct_rejec': b1_correct_reject_2n, 'nbt_b2_2back_correct_rejec': b2_correct_reject_2n,
                     'nbt_b3_2back_correct_rejec': b3_correct_reject_2n, 

                     'nbt_b1_2back_hits_rt': b1_hits_rt_2n, 'nbt_b2_2back_hits_rt': b2_hits_rt_2n,
                     'nbt_b3_2back_hits_rt': b3_hits_rt_2n, 

                     'nbt_b1_3back_hits': b1_hits_3n, 'nbt_b2_3back_hits': b2_hits_3n, 
                     'nbt_b3_3back_hits': b3_hits_3n, 

                     'nbt_b1_3back_false_alarms': b1_false_alarms_3n, 'nbt_b2_3back_false_alarms': b2_false_alarms_3n,
                     'nbt_b3_3back_false_alarms': b3_false_alarms_3n, 

                     'nbt_b1_3back_misses': b1_misses_3n, 'nbt_b2_3back_misses': b2_misses_3n, 
                     'nbt_b3_3back_misses': b3_misses_3n, 

                     'nbt_b1_3back_correct_rejec': b1_correct_reject_3n, 'nbt_b2_3back_correct_rejec': b2_correct_reject_3n,
                     'nbt_b3_3back_correct_rejec': b3_correct_reject_3n,

                     'nbt_b1_3back_hits_rt': b1_hits_rt_3n, 'nbt_b2_3back_hits_rt': b2_hits_rt_3n,
                     'nbt_b3_3back_hits_rt': b3_hits_rt_3n,

                     'nback_nbt_complete': 1}]

            import_data = project.import_records(data)
            print('Participant ' + str(participant_id) + ' data has been added to REDCap.')
```
